# Standard Deviation and Variance

Sure! Here's a **short, quick, and simple** explanation of **Standard Deviation** and **Variance**, with formulas:

---

### 📊 **Variance (σ² or Var)**

* **What is it?**
  It tells you **how spread out** the numbers are from the **mean**.

* **Formula (for population):**

  $$
  \sigma^2 = \frac{1}{N} \sum (x_i - \mu)^2
  $$

* **Words:**
  Average of the **squared differences** from the mean.

---

### 📈 **Standard Deviation (σ)**

* **What is it?**
  It tells you **how much** the numbers typically **deviate** from the mean.

* **Formula (for population):**

  $$
  \sigma = \sqrt{\sigma^2}
  $$

* **Words:**
  It's just the **square root of variance**.

---

### 💡 Simple Example:

Numbers: `2, 4, 6`

* Mean = 4
* Differences from mean = -2, 0, 2
* Squared differences = 4, 0, 4
* **Variance** = (4 + 0 + 4) / 3 = **2.67**
* **Standard Deviation** = √2.67 ≈ **1.63**

---


```python
Sec_A=np.array([75,65,73,68,72,67])
Sec_B=np.array([90,47,43,96,93,51])
y_axis = np.arange(1, 7)
Sec_A_mean = Sec_A.mean()
Sec_A_abs=np.abs(Sec_A-Sec_A_mean)
Sec_A_abs_sq=np.square(Sec_A_abs)
print("Sec A absolute deviation: ", Sec_A_abs)
print("Sec A absolute deviation: ", Sec_A_abs_sq)
Sec_A_Var=np.sum(Sec_A_abs_sq)/len(Sec_A_abs_sq)
print("Sec A Variance: ", Sec_A_Var)
print("Sec A Standard Deviation: ", np.sqrt(Sec_A_Var))

print("Sec B Variance: ", Sec_B.var())
print("Sec B Standard Deviation: ", Sec_B.std())
```

### ✅ Percentile Formula (from Rank)

Percentile = ((N - R) / N) × 100

Where:
- N = Total number of students
- R = Rank of the student (1 is highest)


Sure! Here's a **short and clear version** of **Quartiles** for your notes:

---



#### ✅ What are Quartiles?

* Quartiles **divide sorted data into 4 equal parts**.
* Used to understand **data spread** and find **outliers**.

#### 🔢 Quartile Definitions:

* **Q1 (25th percentile):** 25% of data is below this.
* **Q2 (50th percentile):** Median – 50% of data is below this.
* **Q3 (75th percentile):** 75% of data is below this.

#### 📊 Why Important?

* Helps summarize large datasets.
* Used in **box plots** and **outlier detection**.
* More informative than just mean or average.

#### 🧮 How to Find:

1. Sort the data.
2. Find:

   * Q2 = Median
   * Q1 = Median of lower half
   * Q3 = Median of upper half

#### 📈 Visualization:

Box Plot shows:

* Min, Q1, Q2, Q3, Max
* Helps visualize **spread and symmetry**.



```python
import matplotlib.pyplot as plt
import numpy as np

# Example sorted data
data = [10, 20, 30, 40, 50, 60, 70, 80]
quartiles = np.percentile(data, [25, 50, 75])
Q1, Q2, Q3 = quartiles

# Plotting the data as a box plot
fig, ax = plt.subplots(figsize=(8, 4))
ax.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor='skyblue'))

# Annotating quartiles on the plot
ax.axvline(Q1, color='blue', linestyle='--', label=f'Q1 = {Q1}')
ax.axvline(Q2, color='green', linestyle='--', label=f'Q2 (Median) = {Q2}')
ax.axvline(Q3, color='red', linestyle='--', label=f'Q3 = {Q3}')

# Labels and title
ax.set_title("Understanding Quartiles with a Box Plot")
ax.set_xlabel("Values")
ax.set_yticks([])
ax.legend()

plt.grid(True)
plt.tight_layout()
plt.show()

```
