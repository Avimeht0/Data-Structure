# Standard Deviation and Variance

Sure! Here's a **short, quick, and simple** explanation of **Standard Deviation** and **Variance**, with formulas:

---

###  **Variance ( or Var)**

* **What is it?**
  It tells you **how spread out** the numbers are from the **mean**.

* **Formula (for population):**

  $$
  \sigma^2 = \frac{1}{N} \sum (x_i - \mu)^2
  $$

* **Words:**
  Average of the **squared differences** from the mean.

---

###  **Standard Deviation ()**

* **What is it?**
  It tells you **how much** the numbers typically **deviate** from the mean.

* **Formula (for population):**

  $$
  \sigma = \sqrt{\sigma^2}
  $$

* **Words:**
  It's just the **square root of variance**.

---

###  Simple Example:

Numbers: `2, 4, 6`

* Mean = 4
* Differences from mean = -2, 0, 2
* Squared differences = 4, 0, 4
* **Variance** = (4 + 0 + 4) / 3 = **2.67**
* **Standard Deviation** = 2.67  **1.63**

---


```python
Sec_A=np.array([75,65,73,68,72,67])
Sec_B=np.array([90,47,43,96,93,51])
y_axis = np.arange(1, 7)
Sec_A_mean = Sec_A.mean()
Sec_A_abs=np.abs(Sec_A-Sec_A_mean)
Sec_A_abs_sq=np.square(Sec_A_abs)
print("Sec A absolute deviation: ", Sec_A_abs)
print("Sec A absolute deviation: ", Sec_A_abs_sq)
Sec_A_Var=np.sum(Sec_A_abs_sq)/len(Sec_A_abs_sq)
print("Sec A Variance: ", Sec_A_Var)
print("Sec A Standard Deviation: ", np.sqrt(Sec_A_Var))

print("Sec B Variance: ", Sec_B.var())
print("Sec B Standard Deviation: ", Sec_B.std())
```

##  Percentile Calculation  Summary Notes

###  1. **Percentile Based on Rank (Common in Exams, Education)**

Used to find **what percentile a person falls into** based on their **rank**.

**Formula:**

$$
\text{Percentile} = \left( \frac{N - R}{N} \right) \times 100
$$

**Where:**

* $N$ = Total number of students (or data points)
* $R$ = Rank of the student (1 is top rank)

**Example:**
If there are 50 students and you rank 10th:

$$
\left( \frac{50 - 10}{50} \right) \times 100 = 80\text{th percentile}
$$

---

###  2. **Percentile Value in a Dataset (Using Position/Index)**

---


###  Goal:

Find the **value** in the data that corresponds to a given **percentile (e.g., 25th, 50th, 75th)**.

---

###  **Steps to Calculate Percentile Value**:

1. **Sort the data**.
2. Use the formula to find the **position**:

   $$
   \text{position} = \frac{p}{100} \times (n - 1)
   $$

   * `p` = desired percentile (e.g., 25)
   * `n` = number of data points
3. Split position into:

   * `lower_index = int(position)`
   * `upper_index = lower_index + 1`
   * `weight = position - lower_index`
4. **Interpolate**:

   $$
   \text{value} = data[lower] \times (1 - weight) + data[upper] \times weight
   $$

---

###  Why Use `weight`?

* It tells how far between two data points the percentile lies.
* Example: position = 1.75  weight = 0.75  75% between index 1 and 2.

---

###  Example:

For `data = [10, 20, 30, 40]`, 25th percentile:

1. $\text{position} = 0.25 \times (4 - 1) = 0.75$
2. Between index 0 and 1: (10 and 20)
3. $\text{value} = 10 + 0.75 \times (20 - 10) = 17.5$



---



####  What are Quartiles?

* Quartiles **divide sorted data into 4 equal parts**.
* Used to understand **data spread** and find **outliers**.

####  Quartile Definitions:

* **Q1 (25th percentile):** 25% of data is below this.
* **Q2 (50th percentile):** Median  50% of data is below this.
* **Q3 (75th percentile):** 75% of data is below this.

####  Why Important?

* Helps summarize large datasets.
* Used in **box plots** and **outlier detection**.
* More informative than just mean or average.

####  How to Find:

1. Sort the data.
2. Find:

   * Q2 = Median
   * Q1 = Median of lower half
   * Q3 = Median of upper half

####  Visualization:

Box Plot shows:

* Min, Q1, Q2, Q3, Max
* Helps visualize **spread and symmetry**.



```python
def percentile_rank(data, value):
    data = sorted(data)
    count = sum(1 for v in data if v <= value)
    percentile = (count / len(data)) * 100
    return percentile
```

```python
def calculate_percentile(data, percentile):
    data = sorted(data)
    n = len(data)
    if n == 0:
        raise ValueError("Data list is empty")
    
    # Rank position (zero-based index)
    rank = (percentile / 100) * (n - 1)
    lower_index = int(rank)
    upper_index = min(lower_index + 1, n - 1)
    
    # Linear interpolation
    weight = rank - lower_index
    result = data[lower_index] * (1 - weight) + data[upper_index] * weight
    return result
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example sorted data
# data = [10, 20, 30, 40, 50, 60, 70, 80 ]
data = [ 10, 20, 30, 40, 50, 60, 70, 80 ]
Q1= calculate_percentile(data, 0)
Q2= calculate_percentile(data, 50)
Q3= calculate_percentile(data, 75)

# Plotting the data as a box plot
fig, ax = plt.subplots(figsize=(8, 4))
ax.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor='skyblue'))

# Annotating quartiles on the plot
ax.axvline(Q1, color='blue', linestyle='--', label=f'Q1 = {Q1}')
ax.axvline(Q2, color='green', linestyle='--', label=f'Q2 (Median) = {Q2}')
ax.axvline(Q3, color='red', linestyle='--', label=f'Q3 = {Q3}')

# Labels and title
ax.set_title("Understanding Quartiles with a Box Plot")
ax.set_xlabel("Values")
ax.set_yticks([])
ax.legend()

plt.grid(True)
plt.tight_layout()
plt.show()

```
