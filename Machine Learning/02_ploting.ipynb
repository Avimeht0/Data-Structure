```python
import numpy as np
import matplotlib.pyplot as plt

# Data
Name = np.array(["Arvind", "Bharat", "Chaitanya", "Deepak", "Eshwar", "Feroz", "Gaurav", "Harsh", "Irfan", "Jaspreet"])
Marks = np.array([45, 67, 89, 23, 56, 78, 90, 12, 34, 67])

# Bar plot
plt.bar(Name, Marks)

# Adding title and labels
plt.title("Marks of Students")
plt.xlabel("Name of Students")
plt.ylabel("Marks of Students")

# Rotate x-axis labels to make them readable
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for Class 1 and Class 2
Name = np.array(["Arvind", "Bharat", "Chaitanya", "Deepak", "Eshwar", "Feroz", "Gaurav", "Harsh", "Irfan", "Jaspreet"])
Marks_Class1 = np.array([45, 67, 89, 23, 56, 78, 90, 12, 34, 67])
Marks_Class2 = np.array([55, 60, 80, 40, 50, 85, 75, 20, 30, 60])

# Bar width
bar_width = 0.35

# Positioning for bars (shift by half the bar width for side-by-side comparison)
x = np.arange(len(Name))

# Creating the bars for both classes
plt.bar(x-0.30 , Marks_Class1, bar_width, label="Class 1", color='blue')
plt.bar(x , Marks_Class2, bar_width, label="Class 2", color='orange')

# Adding titles and labels
plt.title("Marks of Students: Class 1 vs Class 2")
plt.xlabel("Name of Students")
plt.ylabel("Marks")
plt.xticks(x, Name, rotation=45)  # Set x-axis labels and rotate them for readability

# Adding a legend to differentiate the classes
plt.legend()

# Show the plot
plt.show()
```
